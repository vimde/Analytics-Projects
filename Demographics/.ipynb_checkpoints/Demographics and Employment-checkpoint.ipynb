#### 19. Which metropolitan area has the highest proportion of interviewees of Hispanic ethnicity? 

** Laredo, TX	**

```python
metropolitanArea = pd.crosstab(index = CPS['MetroArea'], columns=CPS['Hispanic'], margins=True)
#.sort_values(by = 'All')
metropolitanArea = metropolitanArea.rename(columns = {0 : 'Non-Hispanic', 1 : 'Hispanic'})
metropolitanArea['proportion'] = metropolitanArea['Hispanic'] / metropolitanArea['All'] * 100.0
metropolitanArea.sort_values(by = 'proportion')
```

#### 25. What proportion of the interviewees from the "New York-Northern New Jersey-Long Island, NY-NJ-PA" metropolitan area have a country of birth that is not the United States? For this computation, don't include people from this metropolitan area who have a missing country of birth.

** 0.309 or 30.92993% **

```python
interviewees = CPS_Country[CPS_Country['MetroArea'] == 'New York-Northern New Jersey-Long Island, NY-NJ-PA']
totalInterviewees = len(interviewees)
totalIntervieweesNotBornInUnitedStates = len(interviewees[interviewees['Country'] != 'United States']) / totalInterviewees
totalIntervieweesNotBornInUnitedStates * 100.0
```

#### 26. Which metropolitan area has the largest number (note -- not proportion) of interviewees with a country of birth in India?

*New York-Northern New Jersey-Long Island, NY-NJ-PA*

```python
metroAreaWithLargeProportionFromIndia = pd.crosstab(index=CPS_Country['MetroArea'], 
                                                    columns=CPS_Country['Country'] == 'India', margins=True)
metroAreaWithLargeProportionFromIndia = metroAreaWithLargeProportionFromIndia.rename(columns={False :'Other', True : 'India'})
metroAreaWithLargeProportionFromIndia.sort_values('India')
#metroAreaWithLargeProportion.max()
```

#### 27. Which metropolitan area has the largest number (note -- not proportion) of interviewees with a country of birth in Brazil?
